In task 3, we visualize a dataset to see the distribution in it. The dataset that will be used in this task is the Iris dataset from kaggle. 

# IRIS DATA

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['SepalLengthCm'].plot(kind='hist', ec='k',color='y', alpha=0.8)
plt.xlabel('Sepal Length Cm')
plt.title('Histogram of Sepal Length')
```

```python
plt.hist(df['SepalWidthCm'], bins=10, ec='k', color='y', alpha=0.8)
plt.xlabel('Sepal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
plt.show()
```

```python
plt.hist(df['PetalLengthCm'], bins=10, ec='k', color='y', alpha=0.8)
plt.xlabel('Petal Length Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()
```

```python
plt.hist(df['PetalWidthCm'], bins=10, edgecolor='k', color='y', alpha=0.8)
plt.xlabel('Petal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')
plt.show()
```

```python
species_count = df['Species'].value_counts()
df['Species'].value_counts().plot.barh(stacked=True)
plt.xlabel('Frequency')
plt.title('Distrubution of Species')
```
